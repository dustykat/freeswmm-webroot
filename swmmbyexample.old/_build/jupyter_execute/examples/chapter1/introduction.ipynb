# Introduction

## How to Use This Book

This book is a collection of worked SWMM examples from a variety of sources, some real, most are pedagogical.  Most examples will have a set of files needed to replicate the example, you simply have to download the files, and run the SWMM model using the various files to replicate the examples.  Alternatively you can run the models on the FreeSWMM web interface.

## FreeSWMM Web 

The FreeSWMM Web interface is simply a VPS hosted on AWS that is accessed using a web browser.  The browser will connect to a desktop (XFCE) environment which you can use to open SWMM and the associated files. Please remember to disconnect when you are done, and be aware the system is restarted from time to time so ny files you save or damage for that matter are lost.
